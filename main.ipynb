```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df_oscar = pd.read_csv("oscar.csv")
df_genre = pd.read_csv("avg_profit_genre.csv")

# Merge on genre
df_combined = pd.merge(df_genre, df_oscar, on='genre')

# Get unique genres and assign fixed colors
genres = df_combined['genre'].unique()
palette = sns.color_palette("tab10", len(genres))
genre_colors = dict(zip(genres, palette))

# Set style
sns.set(style="whitegrid")

# 1. Bar Plot: Average Profit by Genre
plt.figure(figsize=(10,6))
sns.barplot(
    data=df_combined,
    x='genre',
    y='avg_profit_per_genre',
    palette=genre_colors,
    order=genres
)
plt.title('Average Profit by Genre')
plt.ylabel('Avg Profit ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 2. Bar Plot: Oscar Win Rate by Genre
plt.figure(figsize=(10,6))
sns.barplot(
    data=df_combined,
    x='genre',
    y='oscar_win_rate_percent',
    palette=genre_colors,
    order=genres
)
plt.title('Oscar Win Rate by Genre')
plt.ylabel('Win Rate (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 3. Scatter Plot: Profit vs Oscar Win Rate
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df_combined,
    x='avg_profit_per_genre',
    y='oscar_win_rate_percent',
    hue='genre',
    palette=genre_colors,
    s=100
)
plt.title('Avg Profit vs Oscar Win Rate by Genre')
plt.xlabel('Avg Profit ($)')
plt.ylabel('Oscar Win Rate (%)')
plt.legend(title='Genre')
plt.tight_layout()
plt.show()

# 4. Heatmap: Correlation Between Numeric Fields
plt.figure(figsize=(10,6))
sns.heatmap(df_combined.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Between Metrics')
plt.tight_layout()
plt.show()
```
